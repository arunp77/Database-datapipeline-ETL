```python
import pandas as pd

# Find all the rows in the table
movie_rows = films_imdb.find_all('tr')

# Initialize empty lists to store data
titles = []
years = []

# Iterate through each row to retrieve movie names and years
for row in movie_rows[1:]:
    title_column = row.find('td', class_='titleColumn')
    
    # Check if the title column is found
    if title_column:
        title_element = title_column.find('a')
        
        # Check if the title element is found
        if title_element:
            movie_name = title_element.text.strip()
            titles.append(movie_name)
            
            # Find and extract the release year
            year_element = title_column.find('span', class_='secondaryInfo')
            if year_element:
                release_year = year_element.text.strip('()')
                years.append(release_year)
            else:
                years.append(None)
        else:
            titles.append(None)
            years.append(None)

# Create a DataFrame for movie names and years
df_movie_data = pd.DataFrame({'Movie Name': titles, 'Release Year': years})

# Display the DataFrame
df_movie_data.head()
```

```python
# Find all the rows in the table
movie_rows = films_imdb.find_all('tr')

# Initialize empty lists to store data
titles = []
years = []
imdb_ratings = []

# Iterate through each row to retrieve movie names, years, and IMDb Ratings
for row in movie_rows[1:]:
    title_column = row.find('td', class_='titleColumn')
    
    # Check if the title column is found
    if title_column:
        title_element = title_column.find('a')
        
        # Check if the title element is found
        if title_element:
            movie_name = title_element.text.strip()
            titles.append(movie_name)
            
            # Find and extract the release year
            year_element = title_column.find('span', class_='secondaryInfo')
            if year_element:
                release_year = year_element.text.strip('()')
                years.append(release_year)
            else:
                years.append(None)
            
            # Find and extract IMDb Rating
            rating_column = row.find('td', class_='ratingColumn imdbRating')
            if rating_column:
                imdb_rating_element = rating_column.find('strong')
                if imdb_rating_element:
                    imdb_rating = imdb_rating_element.text.strip()
                    imdb_ratings.append(imdb_rating)
                else:
                    imdb_ratings.append(None)
            else:
                imdb_ratings.append(None)
        else:
            titles.append(None)
            years.append(None)
            imdb_ratings.append(None)

# Create a DataFrame for movie names, years, and IMDb Ratings
df_movie_data = pd.DataFrame({'Movie Name': titles, 'Release Year': years, 'IMDb Rating': imdb_ratings})

# Display the DataFrame
df_movie_data.head()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url_meta = 'https://assets-datascientest.s3.eu-west-1.amazonaws.com/133_webscraping/Metacritic.html'

# Use urlopen from urllib.request
res = urlopen(url_meta)

# Use BeautifulSoup from bs4
soup_meta = BeautifulSoup(res, 'html.parser')

# Print the prettified HTML
print(soup_meta.prettify())
```

- (f). Retrieve all titles and ratings, in a df_meta DataFrame. Display the first lines.

```python
# Find all movie rows
movie_rows = soup_meta.select('table.clamp-list tbody tr')

movie_names = []
metascores = []
user_scores = []

# Iterate through movie rows
for movie_row in movie_rows:
    title_element = movie_row.select_one('td.clamp-summary-wrap h3.title a')
    metascore_element = movie_row.select_one('div.clamp-metascore div.metascore_w')
    user_score_element = movie_row.select_one('div.clamp-userscore div.metascore_w')

    # Check if the elements are found
    if title_element and metascore_element and user_score_element:
        movie_name = title_element.text.strip()
        metascore = metascore_element.text.strip()
        user_score = user_score_element.text.strip()

        movie_names.append(movie_name)
        metascores.append(metascore)
        user_scores.append(user_score)

# Create a DataFrame
df_meta = pd.DataFrame({
    'Movie Name': movie_names,
    'Metascore': metascores,
    'User Score': user_scores
})
```
